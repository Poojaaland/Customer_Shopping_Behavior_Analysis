```python
df["Review Rating"] = df.groupby("Category")["Review Rating"].transform(lambda x: x.fillna(x.median()))
```

```python
df.isnull().sum()
```

```python
df.columns = df.columns.str.lower() #converting the columns name into lowercase
df.columns = df.columns.str.replace(" ","_") # converting the space into underscore 
df = df.rename(columns = {'purchase_amount_(usd)' : "purchase_amount"})
```

```python
df.columns
```

```python
# create a new column (age-group)

labels = ["Young Adult","Adult","Middle-age","Senior"]
df["age_group"] = pd.qcut(df["age"], q=4, labels = labels) 
```

```python
df[["age","age_group"]].head(10)
```

```python
# Create a column (purchase fequency days) 
frequency_mapping = {
    "Fortnightly" : 14,
    "Weekly" : 30,
    "Quarterly" : 90,
    "Bi-weekly" : 14,
    "Annually" : 365,
    "Every 3 months" : 90
}

df["purchase_frequency_days"] = df["frequency_of_purchases"].map(frequency_mapping)
```

```python
df[["purchase_frequency_days","frequency_of_purchases"]].head(10)
```
